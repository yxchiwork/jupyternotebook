# üéµ 0. Initiate Connection String and Set the paths

```python
import requests, os, time, sharepy
from getpass import getpass
from datetime import datetime

sp_path = 'https://hicity.sharepoint.com'
subsite_name = '/sites/test'
site_url = sp_path + subsite_name
upload_to_folder = "/Shared Documents"
upload_from_folder = '/Users/yxchi/Downloads/Docs/Stock'
download_from_folder = '/Shared Documents/test'
download_to_folder = '/Users/yxchi/Downloads'
archive_folder = upload_to_folder + "/Archive/"

if os.path.isfile("sp-session.pkl"):
    s = sharepy.load()
else:
#    s = sharepy.connect(sp_path)
    s = sharepy.connect(sp_path,username='yuxiangchi@hicity.onmicrosoft.com',password='Cs.23311998')
    s.save()
#Update the spliter if system is windows.
if os.name == 'nt':
    spliter = '\\'
else:
    spliter = '/'
```

# ‚è¨ 1. Download

```python
start_time = time.time()
folder = s.get("{}/_api/web/GetFolderByServerRelativeUrl('{}')/Files".format(site_url,subsite_name + download_from_folder)).json()["d"]['results']
i = 0
for f in folder:
    try:
        filelist = sp_path + f['ServerRelativeUrl']
        s.getfile(filelist,filename = download_to_folder + spliter + f['Name'])
        print("File " + f["Name"] + " is downloaded to " + download_to_folder)
        i = i +1
    except Exception as err: 
        print("Some error occurred: " + str(err))
#calculate running time.
elapsed_time = time.time() - start_time
print("There are " + str(i) + " files downloaded in total. \nTime elapsed is: " + time.strftime("%H:%M:%S", time.gmtime(elapsed_time)))
```

# üöÖ 2. Move files üéß  

```python
start_time = time.time()
filelist = s.get("{}/_api/web/GetFolderByServerRelativeUrl('{}{}')/Files".format(site_url,subsite_name,upload_to_folder)).json()["d"]['results']
if filelist:
    for file in filelist:
        try:
            inserttime = datetime.now().strftime("%d-%m-%Y_%I-%M-%S_%p")
            response = s.post("{}/_api/web/getfilebyserverrelativeurl({})/moveto(newurl='{}{}{}', flags=1)".format(site_url,file['ServerRelativeUrl'],site_url,archive_folder,file['Name']))
            print("Moved file " + file['Name'] + " to Archive folder")
        except Exception as err: 
            print("Some error occurred: " + str(err))
    elapsed_time = time.time() - start_time
    print("Moving Files to archive is completed, it took " + time.strftime("%H:%M:%S", time.gmtime(elapsed_time)))
else:
    print("No files in the " + upload_folder + " folder!")
```

# ‚è´ 3. Upload Files

```python
start_time = time.time()
with os.scandir(upload_from_folder) as it:
    for entry in it:
        if not entry.name.startswith('.') and entry.is_file():
            with open(entry, 'rb') as file_input:
                try: 
                    response = s.post("{}/_api/web/GetFolderByServerRelativeUrl('{}')/Files/add(url='{}',overwrite=true)".format(site_url,subsite_name + upload_to_folder,entry.name),data=file_input)
                    print("Uploaded " + entry.name)
                except Exception as err: 
                    print("Some error occurred: " + str(err))
    elapsed_time = time.time() - start_time
    print("Upload is completed, it took " + time.strftime("%H:%M:%S", time.gmtime(elapsed_time)))
```
